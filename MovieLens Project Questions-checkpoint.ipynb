### 8. Visualize how popularity of genres has changed over the years - 10 marks

Note that you need to use the number of releases in a year as a parameter of popularity of a genre

Hint 

1: you need to reach to a data frame where the release year is the index and the genre is the column names (one cell shows the number of release in a year in one genre) or vice versa.
Once that is achieved, you can either use univariate plots or can use the heatmap to visualise all the changes over the years 
in one go. 

Hint 2: Use groupby on the relevant column and use sum() on the same to find out the nuumber of releases in a year/genre.  

```python
# Your answer here
# Unlike questions 4, 5 where I used conditional selection to choose columns of genre
# I have directly provided the list of genre

genre = item[['Action','Adventure','Animation','Childrens','Comedy','Crime','Documentary','Drama','Fantasy','Film-Noir','Horror','Musical','Mystery','Romance','Sci-Fi','Thriller','War','Western','release year']]
genre.head()
```

```python
# each cell represents the number of releases per year for each genre

genre_releases = genre.groupby('release year').sum()
genre_releases
```

```python
# the heatmap shows interesting data
# while understandably, the number of releases for each genre has increased in general more recently
# a significant rise can be observed in movies of genre 'Comedy' and more evidently 'Drama'

plt.figure(figsize=(15,8))
sns.heatmap(genre_releases, cmap='viridis', linecolor='white', linewidths=1)
plt.tight_layout
```

```python
# a few changes to the previous heatmap in terms of cmap and line colour, width

plt.figure(figsize=(15,8))
sns.heatmap(genre_releases, cmap='coolwarm', linecolor='black', linewidths=1)
plt.tight_layout
```

```python
# a clustermap to try and read into the trends of genre popularity

plt.figure(figsize=(15,8))
sns.clustermap(genre_releases, cmap='coolwarm', linecolor='black', linewidths=1)
plt.tight_layout
```

# The clustermap gives a better representation of similar number of releases in certain years and of the genre that have undergone significant change over the years 

### 9. Find the top 25 movies according to average ratings such that each movie has number of ratings more than 100 - 10 marks

Hint : 
1. First find the movies that have more than 100 ratings(use merge, groupby and count). Extract the movie titles in a list.
2. Find the average rating of all the movies and sort them in the descending order. You will have to use the .merge() function to reach to a data set through which you can get the names and the average rating.
3. Use isin(list obtained from 1) to filter out the movies which have more than 100 ratings.

Note: This question will need you to research about groupby and apply your findings. You can find more on groupby on https://realpython.com/pandas-groupby/.

```python
# The data in dataset 'data' have been grouped by column 'movie id' and then .count() method has been applied
# I have merged the resulting dataset with specific columns of dataset 'user'
# Finally I have extracted the movie titles of the movies that had more than 100 ratings from people

# your answer here

merge_data = pd.merge(data.groupby('movie id').count(), item[['movie id','movie title']], on='movie id')
movie_list = merge_data[merge_data['rating'] > 100]['movie title']
movie_list
```

```python
# The data in dataset 'data' have been grouped by column 'movie id' and then .mean() method has been applied
# The resulting dataset has been merged with specific columns of 'user' dataset, on column 'movie id'
# Finally, the movie ratings have been sorted in descending order

top_movies = pd.merge(data.groupby('movie id').mean(), item[['movie id', 'movie title']], on= 'movie id')
top_rated = top_movies.sort_values('rating', ascending= False)
top_rated
```

```python
# Here, the movies that have more than 100 ratings have been sorted into a list starting with the highest rated movies

top_set = top_rated[top_rated['movie title'].isin(movie_list)]
top_set
```

```python
# The top 25 movies on average rating from the list of movies that have a minimum of 100 ratings

top_25 = top_set['movie title'].head(25)
top_25
```
